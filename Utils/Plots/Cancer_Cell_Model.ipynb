```python
# First parse the acquired txt file

# Open the data file;
dfile = open('cancercell.txt','r') 

# Suppose the first line of the file has the metadata
nt =list(dfile.readline())

# Then get all lines from file
with open('cancercell.txt', 'r') as f:
    lines = f.readlines()

"""# Remove the first line of metadata
with open(r'cancercell.txt', 'w') as fp:
    # iterate each line
    for number, line in enumerate(lines):
        # delete line 1
        if number > 0:
            fp.write(line)"""

# Next remove spaces and remove "\n"
lines = [line.replace(' ', '') for line in lines]
lines = [line.rstrip() for line in lines]

# Finally, write lines in the file
with open('cancercell_parsed.txt', 'w') as f:
    f.writelines(lines)

print(lines[99:101])
print(type(lines[0][1]))
print(len(lines))
```

```python
# Then we process the file with matrices of cell status, represented by pre-defined integers;

# Additional metadata or acquired from len(lines[0]);
width = 100
height = 100

# Non-cancer cell is represented by 0, dark green
# Benevolent cancer cell is represented by 1, light green
# Malignant cancer cell is represented by 2, yellow
# Apoptotic cell is represented by 3, red
non, bene, mali, apop = range(4)

# Set Color map
cmap_reversed = matplotlib.cm.get_cmap('RdYlGn_r')

# Initialize a dataframe to store all data
df = pd.DataFrame(columns=['Time', 'Non-Cancerous Cells', 'Benevolent Cells', 'Maliganent Cells', 'Apoptotic Cells'])
```

```python
# Iterate the evolved data and plot the matrices
global time, config
total = width * height
time = 0
for i in range(0, len(lines), 101):
        # Return percentage of different type cells from the initial status
        non_ct = 0
        bene_ct = 0
        mali_ct = 0
        apop_ct = 0
        config = SP.zeros([height, width])
        for x in range(width):
                for y in range(height):
                        if lines[y][x] == '0':
                                non_ct += 1
                                state = non
                                config[y, x] = state
                        elif lines[y][x] == '1':
                                bene_ct += 1
                                state = bene
                                config[y, x] = state
                        elif lines[y][x] == '2':
                                mali_ct += 1
                                state = mali
                                config[y, x] = state
                        elif lines[y][x] == '3':
                                apop_ct += 1
                                state = apop
                                config[y, x] = state                                
        assert non_ct + bene_ct + mali_ct + apop_ct == total, "Number of cells incorrect"
        non_pct = round(non_ct / total * 100, 1)
        bene_pct = round(bene_ct / total * 100, 1)
        mali_pct = round(mali_ct / total * 100, 1)
        apop_pct = round(apop_ct / total * 100, 1)
        new_row = [time, non_pct, bene_pct, mali_pct, apop_pct]
        df.loc[time] = new_row
        print(f"non-cancerous, benevolent, maliganent, apoptotic cell percent are: {non_pct}%, {bene_pct}%, {mali_pct}%, {apop_pct}%.")
        time += 1

        plt.cla()
        plt.pcolor(config, vmin = 0, vmax = 4, cmap = cmap_reversed, label = config)
        plt.axis('image')
        plt.title('t = ' + str(time))
        plt.show()

# Write pandas into file;
dfile1 = open('cancer_cell_log.csv','w') 
df.to_csv('cancer_cell_log.csv', index=False)
```
